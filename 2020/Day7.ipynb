```python
import collections

colors = collections.defaultdict(list)

for line in lines:
    color = re.match(r"(.+) bags contain", line)
    for count, name in re.findall(r"(\d+) (.+?) bags?[,.]", line):
        colors[name].append(color[1])

#colors
```

```python
# Part 1
canHoldGold = set()

def canHoldColor(color):
    for c in colors[color]:
        canHoldGold.add(c)
        #print(c)
        canHoldColor(c)

canHoldColor("shiny gold")

print(len(canHoldGold))
```

```python
# Part 2
subBags = collections.defaultdict(list)
for line in lines:
    color = re.match(r"(.+) bags contain", line)
    for count, name in re.findall(r"(\d+) (.+?) bags?[,.]", line):
        print(name, count)
        subBags[color[1]].append((name, int(count)))

subBags
```

```python
def countSubbags(bagColor):
    count = 0;
    for bag in subBags[bagColor]:
        count = count + bag[1] + bag[1] * countSubbags(bag[0])
    return count

countSubbags("shiny gold")
```
